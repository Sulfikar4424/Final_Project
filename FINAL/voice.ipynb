```python
# <--==== Importing Necessary Libraries ====-->
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
print("Libraries imported")
```

```python
df = pd.read_csv('parkinsons.data')
```

Data Pre-Processing

Seperating Other variables and Target Variables

```python
X = df.drop(columns=['status','name'], axis=1) # Note : dropping column axis = 1; dropping row then axis = 0
Y = df['status']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

Model Training
We will train the model using Support Vector Machine (SVM)

```python
model = svm.SVC(kernel='linear')
```

```python
# training the SVM model with training data
model.fit(X_train, Y_train)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0,1))
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

Model Evaluation

```python
# Accuracy Score on training data
X_train_pred = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_pred)
```

```python
print('Accuracy (Training Data) :', training_data_accuracy*100, '%')
```

```python
# Accuracy Score on test data
X_test_pred = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test, X_test_pred)
```

```python
print('Accuracy (Testing Data) :', testing_data_accuracy*100, '%')
```

Predictive System
We will now test our model taking some input data, since only testing the model is not enough, we also need to check it against some input values given by the user

```python
input_data = (203.18400,211.52600,196.16000,0.00178,0.000009,0.00094,0.00106,0.00283,0.00958,0.08500,0.00468,0.00610,0.00726,0.01403,0.00065,33.04700,0.340068,0.741899,-7.964984,0.163519,1.423287,0.044539)

# changing input data to numpy array
input_data_numpy = np.asarray(input_data)

#reshaping the numpy array 
input_data_reshape = input_data_numpy.reshape(1,-1)

#standardizing the input data 
std_data = scaler.transform(input_data_reshape)

## prediction
prediction = model.predict(std_data)
print(prediction)

if (prediction[0] == 1):
  print('The patient has Parkinson')
elif (prediction[0] == 0):
  print('The patient does not have Parkinson')
else:
  print('Some error in processing')
```
